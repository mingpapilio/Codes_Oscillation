```python
import glob, os
import numpy as np
import scipy.fft as fft
#
a= np.linspace(0,1,num=21)
b= np.linspace(0,1,num=21)
a= np.round(a, 3)
b= np.round(b, 3)
out= np.empty([1,3],dtype=float)
#
for i in range(len(a)):
    if a[i]==1.0 or a[i]==0.0:
        path_a= int(a[i])
    else:
        path_a= a[i]
    for j in range(len(b)):
        if b[j]== 0.0 or b[j]==1.0:
            path_b= int(b[j])
        else:
            path_b= b[j]
        # PLEASE modify the '...' bits to your path to the folder
        focal_path=".../sims/a{}/b{}".format(path_a,path_b)
        os.chdir(focal_path)
        max_amp_log= 0.
        skip_count= 0
        #
        for file in glob.glob("*.txt"):
            if file=="popn.txt":
                test= np.genfromtxt(file, delimiter="\t")
                sample= test[52:202, 3]
                N= len(sample)
                ft= [i / (N/2) for i in fft.fft(sample)]
                if max(np.abs(ft[1:int(N/2)]))> 0.:
                    max_amp_log= max_amp_log+ max(np.abs(ft[1:int(N/2)]))
                else:
                    skip_count= skip_count+ 1
        #
        if skip_count< 1:
            max_amp_log= max_amp_log/(10-skip_count)
        else:
            max_amp_log= np.nan
        out= np.vstack((out,[float(a[i]),float(b[j]),max_amp_log]))
```

```python
out= out[1:,]
# PLEASE modify the '...' bits to your path to the folder
txt_file= open(".../sims/amp.txt", "w")
txt_file.write("a\tb\tamp\n")
for i in range(out.shape[0]):
    txt_file.write("%f\t%f\t%f\n" % (out[i,0],out[i,1],out[i,2]))
# Closing the file
txt_file.close()
```
